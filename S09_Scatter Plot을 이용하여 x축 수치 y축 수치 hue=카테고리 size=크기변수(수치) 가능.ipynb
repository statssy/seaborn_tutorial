```python
### Scatter Plot을 이용하여 x축 수치 y축 수치 hue=카테고리 size=크기변수(수치) 가능
- 스케터 플롯에 x축, y축, category색, 크기에 쓸 변수를 잘 생각하고 그려야한다.
- 카테고리별로 상관관계를 잘 볼 수 있는 그래프
```

* https://seaborn.pydata.org/examples/different_scatter_variables.html

```python
import seaborn as sns
import matplotlib.pyplot as plt # 폰트 등 바꿀때
sns.set(style="whitegrid")
%matplotlib inline
```

```python
# Load the example iris dataset
diamonds = sns.load_dataset("diamonds")
diamonds.shape
```

```python
diamonds.head()
```

```python
# Shift +Tab 키를 누르시면 옵션을 보실 수 있습니다.
plt.figure(figsize=(6.5, 6.5)) # 그래프 크기
clarity_ranking = ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]
sns.scatterplot(data=diamonds, x="carat", y="price", hue="clarity", 
                palette="ch:r=-.2,d=.3_r", # 색상 옵션
                hue_order=clarity_ranking, # 내가 랭킹을 정한걸로 순서대로 표현해달라.
                size="depth", # 사이즈를 어떠한 변수로 크기
                sizes=(1, 8), # 점크기가 1~8까지 나오게(작게 만들어주기 위함) 
                linewidth=0) # 점의 테트리선 굵기
```

```python
# Draw a scatter plot while assigning point colors and sizes to different
# variables in the dataset
f, ax = plt.subplots(figsize=(6.5, 6.5))
sns.despine(f, left=True, bottom=True)
sns.scatterplot(x="carat", y="price",
                hue="clarity", size="depth",
                palette="ch:r=-.2,d=.3_r",
                hue_order=clarity_ranking,
                sizes=(1, 8), linewidth=0,
                data=diamonds, ax=ax)
```
