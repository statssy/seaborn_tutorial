### PairGrid로 한방에 다양한 플롯 보여주기(매우중요)
- pairplot은 평소에 쓰는건데 이렇게 쓰는것도 시각화 보여줄때 괜찮을것 같음. 혹자는 이게 더 별로라는 얘기도 함. 사실 이게 더 조잡한감이 있긴함
- 객체로 공백 그래프를 만들어 놓고 map_lower 등을 이용해서 대각선 기점으로 그림을 그려줌
- 당연히 hue로도 카테고리 색으로 나눠서 볼 수 있음

```python
import seaborn as sns
sns.set(style="white")
```

```python
df = sns.load_dataset("iris")
df.head()
```

```python
g = sns.PairGrid(df, diag_sharey=False) # pairgrid 공백이 나옴
```

```python
g = sns.PairGrid(df, hue="species", diag_sharey=False) # pairgrid 공백이 나옴
g.map(sns.scatterplot)
```

```python
g = sns.PairGrid(df, hue="species", diag_sharey=False) # pairgrid 공백이 나옴
g.map(sns.scatterplot)
g.map_diag(sns.kdeplot)
```

```python
g = sns.PairGrid(df, diag_sharey=False) 
g.map_lower(sns.kdeplot) # 대각선 기점으로 아래
g.map_upper(sns.scatterplot) # 대각선 기점으로 위
g.map_diag(sns.kdeplot, lw=3) # 대각선
```
