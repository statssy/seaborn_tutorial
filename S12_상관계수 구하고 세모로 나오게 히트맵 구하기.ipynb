### 상관계수 구하고 세모로 나오게 히트맵 구하기
- 상관계수 구하기
- 히트맵 그리기
    - 세모로 나오게 mask[np.triu_indices_from(mask)] = True
    - 안에 수치 넣기 annot=True, fmt=".1f"
    - 색깔이 치중되어있을때 중앙이랑 맥스 잡아주기 center=0, vmax=.3

## Plotting a diagonal correlation matrix
* https://seaborn.pydata.org/examples/many_pairwise_correlations.html
* [상관 분석 - 위키백과, 우리 모두의 백과사전](https://ko.wikipedia.org/wiki/%EC%83%81%EA%B4%80_%EB%B6%84%EC%84%9D)
* [Correlation and dependence - Wikipedia](https://en.wikipedia.org/wiki/Correlation_and_dependence)

```python
from string import ascii_letters
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="white")
%matplotlib inline
```

```python
# Generate a large random dataset
rs = np.random.RandomState(33)
d = pd.DataFrame(data=rs.normal(size=(100, 26)),
                 columns=list(ascii_letters[26:]))
d.head(2)
```

```python
# Compute the correlation matrix
corr = d.corr()
corr.head(2)
```

```python
# xy대칭이니까 한쪽만 나오기위한 방법Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
mask[:, 2]
```

```python
# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)
```

```python
plt.figure(figsize=(11, 9)) #그래프 사이즈 크기 좀 크게
sns.heatmap(corr, mask=mask, # 마스크로 반만 나오게
            cmap=cmap, center=0, vmax=.3, # 색이 한쪽으로 치우쳐서 한색만 나오니까 중앙하고 최대를 정해주면 색이 두색으로 나옴
            annot=True, fmt=".1f", # annot : 안에다가 수치를 넣기, fmt : 소수점을 잘라서
            square=True, linewidths=.5, # square랑 linewidths 하면 네모들 사이의 간격
            cbar_kws={"shrink": .5}) # 오른쪽의 컬러바를 좀 세밀?하게 해줌
```

```python
# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
