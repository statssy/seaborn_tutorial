### 박스플롯+분포도 그래프 = 바이올린
- 박스플롯 + 분포도를 같이 보여주자
- x축은 카테고리 y축은 수치로 나누자
- 그룹별로 바로 옆에 그릴수도있고 붙여서 그릴수도있다
- 바이올린 안에 선을 그어서 4분위를 보여줄수도 있고 점을 찍어서 실제데이터를 보여줄수도있다.

## Grouped violinplots with split violins
* https://seaborn.pydata.org/examples/grouped_violinplots.html

```python
import seaborn as sns
sns.set(style="whitegrid", palette="pastel", color_codes=True)

%matplotlib inline
```

```python
# Load the example tips dataset
tips = sns.load_dataset("tips")
tips.info()
```

```python
tips.sample()
```

```python
sns.boxplot(x="day", y="total_bill", data=tips)
```

```python
sns.violinplot(x="day", y="total_bill", data=tips) # x에는 카테고리컬 y에는 수치형을 넣는게 바람직
```

```python
sns.violinplot(x="day", y="total_bill", data=tips, hue="smoker") # hue : 나누기
```

```python
sns.violinplot(x="time", y="tip", data=tips, order=["Dinner", "Lunch"]) # order : 순서 정하기
```

```python
tips["weekend"] = tips["day"].isin(["Sat", "Sun"])
tips.sample()
```

```python
sns.violinplot(x="day", y="total_bill", hue="smoker", split=True, data=tips) # split=True하면 바이올린을 붙여준다
```

```python
# Draw a nested violinplot and split the violins for easier comparison
sns.violinplot(x="day", y="total_bill", hue="smoker",
               split=True, inner="quartile", 
               data=tips)
# inner : quartile 사분위로 선으로 보여주기 다른 것도 있음 점도 있음
sns.despine(left=True) 
```
