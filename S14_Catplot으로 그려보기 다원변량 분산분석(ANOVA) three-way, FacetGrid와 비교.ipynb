### Catplot으로 다원변량 ANOVA three-way 그려보기
- oneway : 카테고리별(독립변수 1개)로 종속변수(1개) 차이 보고싶을때
- twoway : 카테고리별(독립변수 2개)로 종속변수(1개) 차이 보고싶을때(ex. 독립변수 A랑 B랑 바꾸면 어떻게 종속변수가 달라지는데?는 주효과 A*B는 상호작용효과)
- threeway : 독립변수 3개 경우 어쩔수 없이 twoway를 두개 만든꼴로 해야할듯
---
- catplot을 이용해서 옵션으로 kind는 point로, capsize 박스플롯 모자 길이 조정, col로 그림 쪼개기 가능

```python
import seaborn as sns
sns.set(style="whitegrid")
%matplotlib inline
```

```python
# Load the example exercise dataset
df = sns.load_dataset("exercise")
df.shape
```

```python
df.head()
```

```python
df.kind.value_counts()
```

```python
sns.pointplot(data=df, x="time", y="pulse", 
              hue="kind", capsize=.2) # 캡사이즈는 사분위 모자 길이
sns.pointplot(data=df, x="time", y="pulse", 
              hue="kind", ci="sd", capsize=.1) # ci를 차이줘서 변화 볼라고
```

```python
sns.catplot(data=df, x="time", y="pulse", 
            kind="point", # kind를 violin, bar로 그리고 싶으면 바꾸면 됨
            hue="kind", capsize=.2,
            col="diet") # col을 쓰면 그 변수로 아얘 그림을 나눠서 보여줌
```

```python
# Draw a pointplot to show pulse as a function of three categorical factors
g = sns.catplot(x="time", y="pulse", hue="kind", col="diet",
                capsize=.2, palette="YlGnBu_d", height=6, aspect=.75,
                kind="point", data=df)
g.despine(left=True)
```

## 맥박수에 대한 분산과 표준편차 구하기

```python
rest = df.query("kind=='rest'")["pulse"]
rest.var()
```

```python
# 표준편차
rest.var() ** 0.5
```

```python
import numpy as np
np.sqrt(rest.var())
```

```python
walking = df.query("kind=='walking'")["pulse"]
walking.var()
```

```python
np.sqrt(walking.var())
```

```python
running = df.query("kind=='running'")["pulse"]
running.var()
```

```python
np.sqrt(running.var())
```

## 분포를 통해 맥박수 표현하기

```python
sns.kdeplot(rest, label="rest")
sns.kdeplot(walking, label="walking")
sns.kdeplot(running, label="running")
```

## FacetGrid 를 통해 catplot 처럼 그리기

```python
g = sns.FacetGrid(df, col="diet", height=6)
g.map(sns.pointplot, "time", "pulse", "kind")
```
