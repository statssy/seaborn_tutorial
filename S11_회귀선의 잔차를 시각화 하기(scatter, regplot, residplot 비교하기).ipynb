### S11_회귀선의 잔차를 시각화 하기(scatter, regplot, residplot 비교하기)
- scatter로 점찍고 regplot으로 줄긋고 ci도 조정해 보고 residplot으로 잔차 시각화하기

* https://seaborn.pydata.org/examples/residplot.html
* lowess 란? locally weighted robust scatterplot smoothing 의 약자
* 출처 : [lowess - Wiktionary](https://en.wiktionary.org/wiki/lowess)

```python
import numpy as np
import seaborn as sns
sns.set(style="whitegrid")
%matplotlib inline
```

```python
# Make an example dataset with y ~ x
rs = np.random.RandomState(7)
x = rs.normal(2, 1, 75)
y = 2 + 1.5 * x + rs.normal(0, 2, 75)
```

```python
sns.scatterplot(x=x, y=y, color="g")
```

```python
sns.regplot(x=x, y=y, color="g")
```

```python
# 선형으로 그려지지않고 구불구불하게 lowess: locally weighted robust scatterplot smoothing
sns.regplot(x=x, y=y, color="g", lowess=True)
```

```python
# 오차가 얼마만큼 차이나는가? Plot the residuals after fitting a linear model
sns.residplot(x, y, lowess=True, color="g")
```

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(2, 2)
fig.set_size_inches(15, 8)
sns.scatterplot(x, y, color="g", ax=axes[0][0])
sns.regplot(x, y,color="g", ax=axes[0][1])
sns.regplot(x, y, lowess=True, color="g", ax=axes[1][0])
sns.residplot(x, y, lowess=True, color="g", ax=axes[1][1])
```
